```python
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
path = '/content/drive/MyDrive/TA DADANG/Tugas Akhir'
```

```python
  df =pd.read_csv("{}/nr-er.csv".format(path))
df.head()
```

```python
df = df.drop(["Name"], axis=1)
```

```python
df.head(2)
```

```python
df2 =pd.read_csv("drive/My Drive/TA DADANG/Tugas Akhir/label.csv")
print(df2)
```

```python
dataset = pd.concat([df,df2], axis = 1)
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(axis=0, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
label_counts = dataset['NR-ER'].value_counts()

print(label_counts)
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
sel = VarianceThreshold()
```

```python
sel.fit(dataset)
```

```python
sel_fitur = sel.get_feature_names_out()
```

```python
dataset = dataset[sel_fitur]
dataset.head(2)
```

```python
train, test = train_test_split(dataset, test_size=0.20, random_state=42, stratify=dataset['NR-ER'])
```

```python
train.to_csv("{}/train_NRER.csv".format(path), index=None)
```

```python
test.to_csv("{}/test_NRER.csv".format(path), index=None)
```
